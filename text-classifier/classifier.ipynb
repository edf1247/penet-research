```python
import pandas as pd
from keras import models, layers
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import classification_report, roc_auc_score
```

```python
df = pd.read_csv("/users/edfarber/scratch/dataset/multimodalpulmonaryembolismdataset/merged_pe_data.csv")
```

```python
df.shape
```

```python
df = df.drop("pe_type", axis=1)
```

```python
train_df = df.loc[df["split"] != "test"]
```

```python
train_df.shape
```

```python
test_df = df.loc[df["split"] == "test"]
```

```python
test_df.shape
```

```python
x_train = train_df.drop(["idx", "label", "split"], axis=1).to_numpy()
```

```python
x_train.shape
```

```python
y_train = train_df["label"].to_numpy()
```

```python
scaler = StandardScaler()
```

```python
x_train = scaler.fit_transform(x_train)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
clf = LogisticRegression(random_state=42, solver="saga", penalty="l1", C=0.01, max_iter=1000).fit(x_train, y_train)
```

```python
clf.score(x_train, y_train)
```

```python
x_test = test_df.drop(["idx", "label", "split"], axis=1).to_numpy()
y_test = test_df["label"].to_numpy()
```

```python
x_test = scaler.transform(x_test)
```

```python
clf.score(x_test, y_test)
```

```python
y_pred = clf.predict(x_test)
y_prob = clf.predict_proba(x_test)[:, 1]
```

```python
print(classification_report(y_test, y_pred))
```

```python
print(f"AUC-ROC: {roc_auc_score(y_test, y_prob):.3f}")
```
