```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import sklearn.metrics as metrics
```

```python
penet_preds = pd.read_pickle("/users/edfarber/penet-research/penet-model/results/validation_20251129_002646/preds.pickle")
```

```python
val_preds = pd.read_csv("/users/edfarber/penet-research/text-classifier/validation_proba.csv")
```

```python
val_preds
```

```python
for idx in penet_preds:
    penet_preds[idx]["text_pred"] = val_preds.loc[val_preds["idx"] == idx]["text_preds"].to_numpy()[0]
```

```python
penet_preds
```

```python
clean_data = pd.DataFrame.from_dict(penet_preds, orient="index")
```

```python
clean_data.head()
```

```python
correct_label = clean_data.drop(["pred","penet_pred"], axis=1).to_numpy().ravel()
```

```python
text_preds = clean_data["text_pred"].to_numpy()
```

```python
penet_preds = clean_data["pred"].to_numpy()
```

```python
correct_label.shape
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
features = np.column_stack([text_preds, penet_preds])
```

```python
clf = LogisticRegression().fit(features, correct_label)
```

```python
clf.score(features, correct_label)
```

## Testing

```python
# Load the predictions from the penet model and image classifier
image_preds = pd.read_pickle("/users/edfarber/penet-research/penet-model/results/test_with_optuna_20251128_221355/preds.pickle")
text_preds = pd.read_csv("/users/edfarber/penet-research/text-classifier/test_proba.csv")
```

```python
for idx in image_preds:
    image_preds[idx]["text_pred"] = text_preds.loc[text_preds["idx"] == idx]["text_preds"].to_numpy()[0]
```

```python
clean_data = pd.DataFrame.from_dict(image_preds, orient="index")
```

```python
clean_data.head()
```

```python
correct_label = clean_data.drop(["pred","text_pred"], axis=1).to_numpy().ravel()
```

```python
text_preds = clean_data["text_pred"].to_numpy()
```

```python
penet_preds = clean_data["pred"].to_numpy()
```

```python
features = np.column_stack([text_preds, penet_preds])
```

```python
y_prob = clf.predict_proba(features)[:, 1]
```

```python
print(f"AUC-ROC: {metrics.roc_auc_score(correct_label, y_prob):.3f}")
```

## Plotting

```python
fpr, tpr, threshold = metrics.roc_curve(correct_label, y_prob)
roc_auc = metrics.auc(fpr, tpr)
```

```python
import matplotlib.pyplot as plt
plt.title('Receiver Operating Characteristic')
plt.plot(fpr, tpr, 'b', label = 'AUC = %0.2f' % roc_auc)
plt.legend(loc = 'lower right')
plt.plot([0, 1], [0, 1],'r--')
plt.xlim([0, 1])
plt.ylim([0, 1])
plt.ylabel('True Positive Rate')
plt.xlabel('False Positive Rate')
plt.show()
```
